**Cautions Explanation :**

The visible red, green, and blue backgrounds in the sample output are due to the use of the cmap
argument in matplotlib. When displaying the individual color channels (Red, Green, Blue) using
imshow() with specific colormaps such as Reds, Greens, and Blues, matplotlib visualizes each channel
with those distinct color backgrounds to highlight them.
If you were to display the same channels using Pillow (PIL) directly, the output would be in
grayscale, as PIL doesn’t apply the color maps automatically. This is because the channels contain the
intensity values for each color and don’t have the full color representation unless we specifically map
them to RGB or use a colormap as matplotlib does.
Summary:
• Matplotlib with cmap: Displays each channel with its respective color.
• Pillow (PIL): Displays the channels in grayscale, reflecting the intensity of each channel.

# **3. Image Compression and Decompression using PCA.**

1. Load and Prepare Data:

  • Fetch an image of you choice.{If colour convert to grayscale}

  • Center the dataset - Standaridze the Data.

  • Calculate the covaraince matrix of the Standaridze data

```python
data = gray_image_array.copy()
plt.imshow(data, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#Shape of the 'data' arra y to check the dimensions of the image
print(data.shape)  # Output: (height, width) -> (325 rows, 220 columns) for a 256x256 image
```

```python
# Calculate the mean for each column (pixel position across all rows)
mean = np.mean(data, axis=0)

# Center the data by subtracting the mean from each column
centered_data = data - mean

# Check the shape of the centered data
print(centered_data.shape)  # The shape will still be (325, 220)
```

```python
# Calculate the covariance matrix of the centered data
covariance_mat = np.cov(centered_data, rowvar=False)

# Print the shape of the covariance matrix
print(covariance_mat.shape)  # The shape will be (220, 220) for a 220x220 image

# Print the covariance matrix to see how the features change together
print(covariance_mat)
```

2. Eigen Decomposition and Identifying Principal Components:

  • Compute Eigen Values and Eigen Vectors.

  • Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.

  • Identify the Principal Components with the help of cumulative Sum plot.

```python
# Perform Eigenvalue decomposition of the covariance matrix
eigen_values, eigen_vectors = np.linalg.eigh(covariance_mat)

# Sort the indices of eigenvalues in descending order
sorted_indices = np.argsort(eigen_values)[::-1]

# Sort the eigenvalues in descending order
eigenvalues = eigen_values[sorted_indices]

# Sort the eigenvectors corresponding to the highest eigenvalues
eigenvectors = eigen_vectors[:, sorted_indices]
```

```python
# Identify principal components with the help of cumulative sum plot
explained_variance_ratio = eigenvalues/ np.sum(eigenvalues)

# Plot the cumulative explained variance
plt.plot(np.cumsum(explained_variance_ratio))
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

3. Reconstruction and Experiment:

  • Reconstruction: Transform the original data by multiplying it with the selected eigenvectors(PCs) to obtain a lower-dimensional representation.
  • Experiments: Pick Four different combination of principal components with various explained variance value and compare the result.
  • Display the Results and Evaluate.

```python
# Set the number of principal components (k = 35)
k = 35

# Select the first k eigenvectors (top 35 principal components)
components = eigenvectors[:, :k]  # Components correspond to
```

```python
# Project the centered data onto the top k principal components (eigenvectors)
compressed_data = np.dot(centered_data, components)
```

```python
# attempting to build the image with the features with most information only
decompressed_data = np.dot(compressed_data, components.T) + mean
```

```python
#Original image
plt.subplot(1,3,1)
plt.imshow(gray_image_array, cmap ='gray')
plt.title("Original Image")
plt.axis('off')

#Compressed image
plt.subplot(1,3,2)
plt.imshow(compressed_data, cmap='gray', aspect='auto')
plt.title("Compressed Image")
plt.axis('off')

#Decompressed image
plt.subplot(1,3,3)
plt.imshow(decompressed_data, cmap='gray')
plt.title("Decompressed Image")
plt.axis('off')
```

```python
# Define different values of k for experimenting with different numbers of principal components
k1 = 10   # Use the top 10 principal components
k2 = 20   # Use the top 20 principal components
k3 = 50   # Use the top 50 principal components
k4 = 100  # Use the top 100 principal components
k5 = 150  # Use the top 150 principal components

# Select the first k principal components for each experiment
components1 = eigenvectors[:, :k1]  # Top 10 principal components
components2 = eigenvectors[:, :k2]  # Top 20 principal components
components3 = eigenvectors[:, :k3]  # Top 50 principal components
components4 = eigenvectors[:, :k4]  # Top 100 principal components
components5 = eigenvectors[:, :k5]  # Top 150 principal components
```

```python
# Project the centered data onto the top k principal components for each value of k
compressed_data1 = np.dot(centered_data, components1)  # Using the top 10 components
compressed_data2 = np.dot(centered_data, components2)  # Using the top 20 components
compressed_data3 = np.dot(centered_data, components3)  # Using the top 50 components
compressed_data4 = np.dot(centered_data, components4)  # Using the top 100 components
compressed_data5 = np.dot(centered_data, components5)  # Using the top 150 components

# Reconstruct the data from the compressed representation for each value of k
decompressed_data1 = np.dot(compressed_data1, components1.T) + mean  # Reconstruct using the top 10 components
decompressed_data2 = np.dot(compressed_data2, components2.T) + mean  # Reconstruct using the top 20 components
decompressed_data3 = np.dot(compressed_data3, components3.T) + mean  # Reconstruct using the top 50 components
decompressed_data4 = np.dot(compressed_data4, components4.T) + mean  # Reconstruct using the top 100 components
decompressed_data5 = np.dot(compressed_data5, components5.T) + mean  # Reconstruct using the top 150 components
```

```python
# Decompressed images displayed for various numbers of components

plt.subplot(2, 3, 1)  # First subplot (2 rows, 3 columns, first plot)
plt.imshow(decompressed_data1, cmap='gray')  # Display the image with 10 components
plt.title("10 components")
plt.axis('off')

plt.subplot(2, 3, 2)  # Second subplot (2 rows, 3 columns, second plot)
plt.imshow(decompressed_data2, cmap='gray')  # Display the image with 20 components
plt.title("20 components")
plt.axis('off')

plt.subplot(2, 3, 3)  # Third subplot (2 rows, 3 columns, third plot)
plt.imshow(decompressed_data3, cmap='gray')  # Display the image with 50 components
plt.title("50 components")
plt.axis('off')

plt.subplot(2, 3, 4)  # Fourth subplot (2 rows, 3 columns, fourth plot)
plt.imshow(decompressed_data4, cmap='gray')  # Display the image with 100 components
plt.title("100 components")
plt.axis('off')

plt.subplot(2, 3, 5)  # Fifth subplot (2 rows, 3 columns, fifth plot)
plt.imshow(decompressed_data5, cmap='gray')  # Display the image with 150 components
plt.title("150 components")
plt.axis('off')

# Tight layout to ensure that the plots are spaced correctly
plt.tight_layout()

# Show the plots
plt.show()
```

Initially, we identified the principal components by sorting the eigenvectors based on their corresponding eigenvalues in descending order. The top eigenvectors represent the principal components, which capture the most significant variance in the image.

To determine how many principal components are needed for reconstructing the image, we analyzed the cumulative explained variance plot.

By experimenting with various numbers of components to reconstruct the image, we can observe what happens when we use different numbers of components.

10 components: The image is a rough outline with only the basic information.

20 components: The image is blurry, but you can still make out the main objects.

50 components: The image looks very close to the original, with most details preserved.

100 components: The reconstructed image is almost identical to the original.

150 components: Adding more components doesn’t improve the image much.

In conclusion, PCA helps reduce image size while keeping important details. Using 50 to 100 components is enough to get a good reconstruction without much loss of quality.
